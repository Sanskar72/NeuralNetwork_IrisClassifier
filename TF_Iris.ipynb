```python
#IMPORT LIBRARIES
import pandas as pd
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
```

```python
df = pd.read_csv('Iris.csv')
```

```python
#USING PD.FACTORIZE()
encoded_data, mapping_index = pd.Series(df['Species']).factorize()
```

```python
print(encoded_data)
print(mapping_index)
```

```python
encoded_data.shape
```

```python
#Creating the label and the feature sets
X = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y = encoded_data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=19, shuffle = True, test_size = 0.2)
```

```python
#DEFINE THE NN MODEL
model = tf.keras.Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal', input_shape=(4,)))
model.add(Dense(8, activation='relu', kernel_initializer='he_normal'))
model.add(Dense(3, activation='sigmoid'))
```

```python
# compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# fit the model
model.fit(X_train, y_train, epochs=150, batch_size=32, verbose=1)
```

```python
# evaluate the model
loss, acc = model.evaluate(X_test, y_test, verbose=1)
print('Test Accuracy: %.3f' % acc)
print('Test Loss: ', loss)
```

```python
# make a prediction
yhat = model.predict(X_test)
ind = np.argmax(yhat, axis = 1)
j=0
for i in ind:
  print(mapping_index[i], i)
  print(y_test[j])
  j=j+1
```
